#### 定义一个学生类，有以下属性:
- 姓名
- 年龄
- 成绩（语文，数学，英语）每课成绩类型为整数 类方法
- 获取学生的姓名，get_name()返回类型，str
- 获取学生的年龄，get_age()返回类型，int
- 返回3门科目中最高的分数，get_course()，int

```python
class Student():
    def __init__(self, name, age, scores):
        self.name = name
        self.age = age
        self.scores = scores
        
    def get_name(self):
        return self.name
    
    def get_age(self):
        return self.age
    
    def get_course(self):
        return max(self.scores)
    

s = Student('liliu', 18, [20, 48,99])
print(s.name)   
print(s.age)
print(s.scores)
print("*" * 40)
print(s.get_name())
print(s.get_age())
print(s.get_course())
```

#### 定义一个字典类DictClass，完成如下功能
- 删除某个key del_dict(key)
- 判断某个键是否在字典里，如果在返回对应的值，不在则返回'not found' get_dict()
- 返回键组成的列表，返回类型：list get_key()
- 合并字典，并且返回合并后字典的values组成的列表，返回类型：list update_dict()

```python
class DictClass():
    
    def __init__(self, dict):
        self.dict = dict
        
    def del_dict(self, key):
        # 判断key是否在字典中
        if key not in self.dict.keys():
            return 'key不在字典中'
        else:
            self.dict.pop(key)
            return '删除成功'
     
    def get_dict(self, key):
        if key not in self.dict.keys():
            return 'not found'
        else:
            return self.dict[key]
        
    def get_key(self):
        return self.dict.keys()
    
    def update_dict(self, dict2):
        self.dict = dict(self.dict, **dict2)
        return self.dict.values()
    
d = DictClass({'a':1, 'b':2, 'c':3})
print(d.__dict__)
print(d.del_dict('c'))
print(d.get_dict('c'))
print(d.get_dict('a'))
print(d.get_key())
print(d.update_dict({'d':4}))
print(d.__dict__)
```

#### 定义一个列表操作类，Listinfo
- 包括的方法
    - 列表元素添加，add_key()，输入的必须是字符串或者数字
    - 列表元素取值，get_key()
    - 列表合并，update_list(list)
    - 删除并返回最后一个元素，del_key()
    

```python
class Listinfo():
    
    def __init__(self, list):
        self.list = list
        
    def add_key(self, element):
        # 判断输入的是否是字符串或数字
        if isinstance(element,(str, int)):
            self.list.append(element)
            return self.list
        return '只能添加数字或字符串'
    
    def get_key(self, index):
        # 判断索引是否list范围内
        if index >= 0 and index < len(self.list):
            return self.list[index]
        else:
            return '索引超出范围了'
    def update_list(self, list2):
        self.list.extend(list2)
        return self.list
    
    def del_key(self):
        # 判断列表是否为空
        if len(self.list) > 0:
            return self.list.pop(-1)
        else:
            return "列表是空的"
        
l = Listinfo([1,2,3,4,5,6])
print(l.__dict__)
print(l.add_key(7))
print(l.get_key(0))
print(l.update_list([8,9,10]))
print(l.del_key())
```

#### 定义一个集合操作类
- 包括的方法
    - 集合元素添加：add_setinfo()
    - 集合的交集：get_intersection()
    - 集合的并集：get_union()
    - 集合的差集：get_difference()

```python
class SetInfo():
    
    def __init__(self, section):
        self.sect = section
        
    def add_setinfo(self, element):
        self.sect.add(element)
        return self.sect
    
    def get_intersection(self, sect2):
        if isinstance(sect2, set):
            return self.sect & sect2
        return '你传入的不是set'
    
    def get_union(self, sect2):
        if isinstance(sect2, set):
            return self.sect | sect2
        return '你传入的不是set'
    
    def get_difference(self, sect2):
        if isinstance(sect2, set):
            return self.sect - sect2 
        return '你传入的不是set'
    
A = {2,3,4,5,6,8,9}
B = {1,3,5,7}
s = SetInfo(A)
print(s.add_setinfo(10))
print(s.get_intersection(B))
print(s.get_union(B))
print(s.get_difference(B))
```

#### 定义一个名票系统
- 门票的原价是100元
- 当周末的时候门票涨价20%
- 小孩半价
- 计算另个成年人和一个小孩的平日票价

```python
class Ticket():
    def __init__(self, weekend=False, child=False):
        self.price = 100
        
        if weekend == True:
            self.inc = 1.2
        else:
            self.inc = 1
        if child == True:
            self.disc = 0.5
        else:
            self.disc = 1
            
    def cat_price(self, num):
        self.p = self.price * self.inc * self.disc * num
        return self.p

a = Ticket()
c = Ticket(child=True)

print("两个男人和一个小孩的票价为{}".format(a.cat_price(2)+c.cat_price(1)))
```

#### 游戏编程 按以下要求定义一个乌龟类和鱼类并尝试编程
- 假设游戏场景为(x,y), 0<=x<=10, 0<=y<=10
- 游戏生成1只乌龟和10条鱼
- 他们的移动方向均随机
- 乌龟的最大移动能力是2（乌龟可以随机选择移动是1还是2），鱼的最大移动能力是1
- 当移动到场景边缘，自动向反方向移动
- 乌龟初始化体力为100（上限）
- 乌龟每移动一次，体力消耗1
- 当乌龟和鱼重叠时，乌龟吃掉鱼，乌龟体力+20
- 鱼不计算体力
- 当乌龟体力为0或者鱼的数量为0时，游戏结束！

```python
import random as r

class Turtle():
    def __init__(self):
        self.power = 100 # 乌龟体力有初始值
        
        # 初始化乌龟的位置
        self.x = r.randint(0,10)
        self.y = r.randint(0,10)
        
    def move(self):
        new_x = r.choice([1, 2, -1, -2]) + self.x
        new_y = r.choice([1, 2, -1, -2]) + self.y
        
        # 判断乌龟的移动位置是否超出边界
        if new_x < 0:
            self.x = 0 - (new_x - 0)
        elif new_x > 10:
            self.x = 10 - (new_x - 10)
        else:
            self.x = new_x
            
        if new_y < 0:
            self.y = 0 - (new_y - 0)
        elif new_y > 10:
            self.y = 10 - (new_y - 10)
        else:
            self.y = new_y
        
        self.power -= 1
        return (self.x, self.y)# 一定要有返回，不然没有结果
    
    def eat(self):
        self.power += 20
        if self.power > 100:
            self.power = 100
        
class Fish():
    def __init__(self):
    
        self.x = r.randint(0,10)
        self.y = r.randint(0,10)
        
    def move(self):
        new_x = r.choice([1, -1]) + self.x
        new_y = r.choice([1, -1]) + self.y
        
        if new_x < 0:
            self.x = 0 - (new_x - 0)
        elif new_x > 10:
            self.x = 10 - (new_x - 10)
        else:
            self.x = new_x
            
        if new_y < 0:
            self.y = 0 - (new_y - 0)
        elif new_y > 10:
            self.y = 10 - (new_y - 10)
        else:
            self.y = new_y
            
        return (self.x, self.y) # 一定要有返回，不然没有结果   

turtle = Turtle()
fish = []
for i in range(10):
    f = Fish()
    fish.append(f)

while True:
    if len(fish) == 0:
        print("鱼被吃光了，游戏结束")
        break
    if turtle.power <= 0:
        print("乌龟体力耗尽了，游戏结束")
        break
    
    pos = turtle.move()
    
    for f in fish[:]:# 把列表拷贝一份传给迭代器，避免直接修改原列表造成其他问题
        if f.move() == pos:
            turtle.eat()
            fish.remove(f)
            print("有一条鱼被吃掉啦")
```

#### 定义一个点(point)和直线(Line)，使用getlen方法获取两点构成直线的长度

```python
import math

class point():
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        
    def get_x(self):
        return self.x
    
    def get_y(self):
        return self.y
    
    
class Line():
    def __init__(self, p1, p2):
        self.x = p1.get_x() - p2.get_x()
        self.y = p1.get_y() - p2.get_y()
        
    def get_len(self):
        self.len = math.sqrt(self.x * self.x + self.y * self.y)
        return self.len
    
    
p1 = point(2,3)
p2 = point(5,7)

l = Line(p1,p2)
print(l.get_len())
```
