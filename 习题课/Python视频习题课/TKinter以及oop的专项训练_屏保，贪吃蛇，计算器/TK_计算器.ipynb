```python
# 第三步，添加逻辑功能
from tkinter import *
from decimal import Decimal

root = Tk()
# 定义面板大小
root.geometry('300x380')
root.title("北京图灵计算器")

# 定义面板
# bg代表背景颜色，，#ddddddd是十六进制表示颜色的一个串
frame_show = Frame(width=350, height=150, bg="#cccccc")
frame_show.pack()

# 定义顶部区域
sv = StringVar()
sv.set("0")

# anchor: 定义控件的锚点，e代表右（东）
show_label = Label(frame_show, textvariable=sv, bg="gray", width=15, height=1, \
                   font=("黑体", 20, 'bold'), justify=LEFT, anchor='e')
show_label.pack(padx=10, pady=10)


'''
考虑以下几种情况：
1. 按下数字
2. 按下操作符号
3. 只考虑只靠虑两个数字的操作运算，不考虑多个数的操作运算
'''
num1 = ''
num2 = ''
operator = ''
whole = '' 

def delete():
    global num1, num2, operator, whole
    if whole != '':
        if whole[-1].isdigit():
            if whole == num1:
                num1 = num1[0:len(num1)-1]
            else:
                num2 = num2[0:len(num2)-1]
        whole = whole[0:len(whole)-1]
        sv.set(whole)
    
def clear():
    global num1, num2, operator, whole
    num1 = ''
    num2 = ''
    operator = ''
    whole = '' 
    sv.set("0")
    
def fan():
    global num1, num2, operator, whole
    if whole != '':
        if whole[-1].isdigit() and num2 == '':
            if whole[0] == '-':
                num1 = str(abs(float(num1)))
                whole = num1
            else:
                num1 = str(float(num1)*(-1))
        elif whole[-1].isdigit() and num2 != '':
            Num2 = float(num2)*(-1)
            whole = whole.replace(num2, Num2)
            num2 = Num2
        sv.set(whole)
        
def ce():
    sv.set("0")



def change(num):
    '''
    按下一个数字需要考虑两种情况
    1. 数字属于第一个操作数
    2. 数字属于第二个操作数
    3. 判断是否属于第一个操作数，可以通过operator判断
    '''
    global num1, num2, operator, whole
    # 假如操作符operator是None，表明肯定是第一个操作数
    if operator == '=':
        if num != '.':
            operator == ''
            num1 = str(num)
            whole = num1
            sv.set(whole)
    elif operator == '':
        if num == '.' and num1 == '':
            pass
        elif num == '.' and '.' in num1:
            pass
        else:
            num1 = num1 + str(num)
            whole = whole + str(num)
            #如果是第一个操作数，则显示第一个操作数
            sv.set(whole)
        sv.set(num1)
    else:
        if num == '.' and num2 == '':
            pass
        elif num == '.' and '.' in num2:
            pass
        else:
            num2 = num2 + str(num)
            whole = whole + str(num)
            # 如果是第二个操作数，则显示完整的式子
            sv.set(whole)
        
def res(s):
    if s[0] == '-':
        for i in s[1:len(s)]:
            if i in ['+','-','*','/']:
                op = i
                li = s.split(op)
                if s.count('-') == 3:
                    Al = i + 'Decimal(li[1])' + i + i + 'Decimal(li[3])'
                elif s.count('-') == 2 and i == '-':
                    Al = i + 'Decimal(li[1])' + i + 'Decimal(li[2])'
                else :
                    Al = 'Decimal(li[0])' + i + 'Decimal(li[1])'
                return(eval(Al))
                break
    else:
        for i in s:
            if i in ['+','-','*','/']:
                op = i
                li = s.split(op)
                if s.count('-') == 2:
                    Al = 'Decimal(li[0])' + i + i + 'Decimal(li[2])'
                else :
                    Al = 'Decimal(li[0])' + i + 'Decimal(li[1])'
                return(eval(Al))
                break

def operation(op):
    global num1, num2, operator, whole
    if op == '-' and num1 == '':
        whole = whole + op
        sv.set(op)
        print(whole)
    elif num1 != '' and whole[-1].isdigit() and op in ['+','-','×','÷'] and num2 == '':
        operator = op
        whole = whole + op
        sv.set(whole)
        print(whole)
    elif op == '-' and whole[-2].isdigit() and whole[-1] in ['+','-','×','÷'] and num2 == '':
        operator = op
        whole = whole + op
        sv.set(whole)
        print(whole)
    elif num2 !='' and num2[-1].isdigit() and op in ['+','-','×','÷']:
        operator = op
        num1 = str(eval("res(whole)"))
        whole = num1 + operator
        sv.set(whole)
        print(whole)
        num2 = ''
    elif whole != '' and op == '=' and whole[-1].isdigit():
        if operator != '=':
            operator = op
            num1 = str(eval("res(whole)"))
            whole = num1
            sv.set(whole)
            print(whole)
            num2 = ''
        
# 按键区域
frame_bord = Frame(width=400, height=350, bg="#cccccc")

b_del = Button(frame_bord, text='←', width=4, height=1, command=delete)
b_del.grid(row=0, column=0)
b_clear = Button(frame_bord, text='C', width=4, height=1, command=clear)
b_clear.grid(row=0, column=1)
b_fan = Button(frame_bord, text='±', width=4, height=1, command=fan)
b_fan.grid(row=0, column=2)
b_ce = Button(frame_bord, text='CE', width=4, height=1, command=ce)
b_ce.grid(row=0, column=3)

b_1 = Button(frame_bord, text='1', width=4, height=2, command=lambda:change("1"))
b_1.grid(row=1, column=0)
b_2 = Button(frame_bord, text='2', width=4, height=2, command=lambda:change("2"))
b_2.grid(row=1, column=1)
b_3 = Button(frame_bord, text='3', width=4, height=2, command=lambda:change("3"))
b_3.grid(row=1, column=2)
b_4 = Button(frame_bord, text='4', width=4, height=2, command=lambda:change("4"))
b_4.grid(row=2, column=0)
b_5 = Button(frame_bord, text='5', width=4, height=2, command=lambda:change("5"))
b_5.grid(row=2, column=1)
b_6 = Button(frame_bord, text='6', width=4, height=2, command=lambda:change("6"))
b_6.grid(row=2, column=2)
b_7 = Button(frame_bord, text='7', width=4, height=2, command=lambda:change("7"))
b_7.grid(row=3, column=0)
b_8 = Button(frame_bord, text='8', width=4, height=2, command=lambda:change("8"))
b_8.grid(row=3, column=1)
b_9 = Button(frame_bord, text='9', width=4, height=2, command=lambda:change("9"))
b_9.grid(row=3, column=2)
b_0 = Button(frame_bord, text='0', width=4, height=2, command=lambda:change("0"))
b_0.grid(row=4, column=0)
b_p = Button(frame_bord, text='.', width=4, height=2, command=lambda:change("."))
b_p.grid(row=4, column=1)

b_add = Button(frame_bord, text='+', width=4, height=2, command=lambda:operation("+"))
b_add.grid(row=1, column=3)
b_sub = Button(frame_bord, text='-', width=4, height=2, command=lambda:operation("-"))
b_sub.grid(row=2, column=3)
b_mul = Button(frame_bord, text='x', width=4, height=2, command=lambda:operation("*"))
b_mul.grid(row=3, column=3)
b_div = Button(frame_bord, text='÷', width=4, height=2, command=lambda:operation("/"))
b_div.grid(row=4, column=3)
b_equ = Button(frame_bord, text='=', width=4, height=2, command=lambda:operation("="))
b_equ.grid(row=4, column=2)

frame_bord.pack(padx=10, pady=10)

root.mainloop()
```
