### 函数的参数和返回值
- 参数：负责给函数传递一些必要的数据或者信息
    - 形参（形式参数）：在函数定义的时候用到的参数没有具体的值，只是一个占位的符号，称为形参
    - 实参（实际参数）：在调用函数的时候输入的值
- 返回值：调用函数时的一个执行结果
    - 使用return关键字返回结果
    - 如果没有return，默认返回一个None
    - 函数一旦执行return语句，则无条件返回，即结束函数的执行

```python
# 形参和实参的案例
# 参数person只是一个符号，代表的是调用的时候的某一个数据
# 调用的时候用另一个

def hello(person):
    print("{0}，干甚呢".format(person))
    print("我要起飞了")

p = "小郑"
# 调用函数，需要吧p作为实参传入
hello(p)
```

```python
# 九九乘法表
# version 1.0

for o in range(1,10): # 控制外循环，从1 到 9
    # 打印一行
    for i in range(1,o + 1): # 内循环，每次从第一个数字开始，打印到跟行数一致结束
        if (o == 3 and i == 3) or (o == 4 and i == 3): # 添加空格
            print(" ",end="")
                
        # print函数默认任务打印完毕后换行（内建函数！！！）
        print(o * i,end = " ")
    print()
```

```python
# 定义一个函数，打印九九乘法表
# version 2.0
def jiujiu():
    for o in range(1,10): # 控制外循环，从1 到 9
    # 打印一行
        for i in range(1,o + 1): # 内循环，每次从第一个数字开始，打印到跟行数一致结束
            print(o * i,end = " ")
        print()
        
    return None


jiujiu()
print()
jiujiu()
```

```python
# 改造上面函数
# version 3.0
def printline(line_num):
    '''
    line_num:代表行号
    
    打印一行九九乘法表
    '''
    for i in range(1,line_num + 1):
        print(line_num * i,end="  ")
    print()    
    return None

def jiujiu():
    for o in range(1,10): # 控制外循环，从1 到 9
        printline(o)
        
    return None

jiujiu()
```

## 收集参数
- 把没有位置，不能定义时的参数位置对应的参数，放入一个特定的数据结构中
- 语法

        def func(*args):
            func_body
            按照list使用方式访问args得到传入的参数
            
        调用：
        func(p1,p2,p3......)
- 参数名args不是必须这么写，但是，我们推荐直接用args，约定俗称
- 参数名args前需要有星号*
- 收集参数可以和其他参数共存

```python
# 收集参数代码
# 函数模拟一个学生进行自我介绍，但具体内容不清楚
# args把他看做一个list
def stu(*args):
    print("Hello, 大家好, 我自我为介绍一下, 简单说两句：")
    # type函数作用是检测变量的类型
    print(type(args))
    for item in args:
        print(item)
        

stu("liuliu", 18, "tuling", "wudi", "single")

stu("小枫灿")
```

```python
# 收集参数可以不带任何实参调用，此时收集参数为空tuple
stu()
```

```python
# 如果使用关键字参数格式调用，会出现问题
stu(name="liulinrui")
```

### 收集参数之关键字收集参数
- 把关键字参数按照字典的格式存入收集参数
- 语法：

        def func( **kwargs):
            func_body
            
            
        # 调用
        func(p1=v1,...)
        
        - kwargs一般为约定俗称
        - 调用的时候，把多余的关键字参数放入kwargs
        - 访问kwargs需要按照字典格式访问

```python
# 收集参数之关键字收集参数案例
# 自我介绍
# 调用的时候需要用到关键字参数调用
def stu(**kwargs):
    # 在函数体内对于kwargs的使用不用带星号
    print("Hello, 大家好，我自我为介绍一下, 简单说两句：")
    print(type(kwargs))
    # 对于字典的访问，py2和py3有区别
    for k,v in kwargs.items():
        print(k, "---", v)
    
stu(name="fengcan", age=19, addr="beijin", love="wangerqing")

print("* " * 30)


stu(name="枫灿")
```

```python
# 收集参数为空案例
stu()
```

### 收集函数混合调用的问题
- 收集参数，关键字参数，普通参数可以混合使用
- 使用规则就是，普通和关键字参数优先
- 定义的时候按普通，关键字，收集参数tuple，收集参数dict

```python
# 混合调用案例
# stu 模拟一个学生的自我介绍
def stu(name, age, *args, hobby="没有", **kwargs):
    print("Hello, 大家好")
    print("我叫{0}，我今年{1}岁了。".format(name, age))
    if hobby == "没有":
        print("我没有爱好，很遗憾")
    else:
        print("我的爱好是{0}".format(hobby))
    print("* " * 30)
    
    for i in args:
        print(i)
    
    print("# " * 30)
    
    for k,v in kwargs.items():
        print(k, "-----", v)
        
        
# 开始调用

name = "fengwang"
age = 25

# 调用的不同格式
stu(name, age)



```

# 变量作用域
- 变量由作用范围限制
- 分类：按照作用域分类
    - 全局（global）：在函数外部定义
    - 局部（local）：在函数内部定义
- 变量的作用范围：
    - 全局变量：在整个全局范围内都有效
    - 全局变量可以在局部范围内可以使用
    - 局部变量在局部范围内可以使用，其他范围不行
- LEGB原则
    - L：(Local)局部作用域
    - E: (Enclosing function locale)外部嵌套函数作用域
    - G：(Global module) 函数定义所在模块作用域
    - B: (Buildin) Python内置模块的作用域

```python
# 认为a1是全局的
a1 = 100

def fun():
    print(a1)
    print("I am in fun")
    # a2的作用范围是fun
    a2 = 99
    print(a2)
    
    
print(a1)
fun()

print(a2)    
```

# 递归函数
- 函数直接或者间接调用自身
- 优点：简洁，理解容易
- 缺点：对递归深度有限制，消耗资源大
- Python对递归深度有限制，超过限制报错
- 在写递归程序的时候，一定注意结束条件


```python
# 递归调用深度限制代码

x = 0

def fun():
    global x
    x += 1
    print(x)
    # 函数自己调用自己
    fun()
    
# 调用函数
fun()
```

# 斐波那契数列
- 一列数字，第一个值是1，第二个值是1，从第三个开始，每一个数字的值等于前两个数字出现的值的和
-  数学公式为：f(1) = 1, f(2) = 1, f(n) = f(n-1) + f(n-2) 
-  例如：1，1，2，3，5，8，13，21.....

```python
# ×××下面求斐波那契数列函数有一定问题，比如n一开始就是负数，如何修正×××
# n表示求第n个数字的斐波那契数列的值
def fib(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    # 思考：为什么后面 return 能够正确执行，而不用else语句
    return fib(n-1) + fib(n-2)
    
print(fib(3))
```

```python
print(fib(5))
```

```python
print(fib(10))
```

```python
# 修正斐波那契数列！！！
def fib(n):
    if n <= 0:
        return("出错")
    if n == 1:
        return 1
    if n == 2:
        return 1
    # 思考：为什么后面 return 能够正确执行，而不用else语句
    return fib(n-1) + fib(n-2)
    
print(fib(3))
```

# 汉诺塔问题
- 有三个塔，分别是A，B，C三个塔。A塔上面分别从小到大、从上到下有n个盘子(B，C塔上没有！！！)，要求把A塔的盘子移动到C塔上，规则如下：
    - 每次只能移动一个盘子。
    - 任何时候、任何一次的移动，三个塔的状态必须全都是小盘子在上，大盘子在下。


## 方法：
- n = 1，A -> C
- n = 2, A -> B, A -> C, B -> C
- n = 3, A上两个盘子通过C移动到B上，调用递归实现，然后把A最大盘子移到C上，最后再把B上的盘子借助A移动到C上，调用递归实现
- n = n, A上n-1个盘子通过C移动到B上，调用递归实现，然后把A最大盘子移到C上，最后再把B上的盘子借助A移动到C上，调用递归实现


```python
def hano(n, a, b, c):
    '''
    汉诺塔的递归实现
    n: 代表几个盘子
    a: 代表第一个塔
    b: 代表第二个塔
    c: 代表第三个塔
    '''
    if n == 1:
        print(a, "-->", c)
        return None
    '''
        if n == 2:
        print(a, "-->", b)
        print(a, "-->", c)
        print(b, "-->", c)
        return None
    '''
    # 把n-1个盘子，从a塔借助c塔挪到b塔上
    hano(n-1,a, c, b)
    print(a, "-->", c) # a上最大盘子移到c上
    # 把n-1个盘子，从b塔借助a塔挪到c塔上
    hano(n-1,b, a, c)
a = "A"
b = "B"
c = "C"


n = 1
hano(n, a, b, c)
        
```
